```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Replace 'dataset.csv' with your actual filename
df = pd.read_csv("Road Accident Data.csv")
df.head()  # Display first few rows
df.info()  # Check data types and missing values
```

Q1. Frequency of Accidents Over Time

```python
# Convert date column to datetime (Assuming 'date' is the column name)
df['Accident Date'] = pd.to_datetime(df['Accident Date'])

# Extract time-based features
df['Year'] = df['Accident Date'].dt.year
df['Month'] = df['Accident Date'].dt.month
df['Day_of_Week'] = df['Accident Date'].dt.day_name()
df['Time'] = df['Accident Date'].dt.time

# Total number of accidents
total_accidents = len(df)
print("Total accidents:", total_accidents)

# Plot accidents by year
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Year')
plt.title("Accidents per Year")
plt.show()

# Accidents by day of the week
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Day_of_Week', order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title("Accidents by Day of the Week")
plt.show()

# Accidents by hour of the day
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Time')
plt.title("Accidents by Hour of the Day")
plt.show()
```

Findings from Q3:This will show peak accident times.

Q2.  Geographical Distribution

```python
# Top accident locations
df['Local_Authority_(District)'].value_counts().head(10).plot(kind='barh')
plt.title("Top 10 Locations with Most Accidents")
plt.xlabel("Number of Accidents")
plt.ylabel("Location")
plt.show()
```

Findings from Q2: This will highlight accident hotspots.

Q3.  Accident Severity Analysis

```python
# Count of accident severity
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Accident_Severity', order=df['Accident_Severity'].value_counts().index)
plt.title("Accident Severity Distribution")
plt.show()

# Percentage of fatalities and serious injuries
severity_counts = df['Accident_Severity'].value_counts(normalize=True) * 100
print(severity_counts)
```

Findings from Q3: This will show the distribution of accident severity.

Q4. Demographic Insights

```python
# Age distribution of individuals in accidents
plt.figure(figsize=(10, 5))
sns.histplot(df['age'], bins=20, kde=True)
plt.title("Age Distribution of Individuals in Accidents")
plt.show()

# Gender distribution
plt.figure(figsize=(6, 5))
sns.countplot(data=df, x='gender')
plt.title("Gender Distribution in Accidents")
plt.show()
```

 Findings from Q4: Helps determine which age group is more involved.

Q5. Environmental and Road Conditions

```python
# Weather conditions vs. accident frequency
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Weather_Conditions', order=df['Weather_Conditions'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Accidents by Weather Conditions")
plt.show()

# Accidents by road type
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Road_Type', order=df['Road_Type'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Accidents by Road Type")
plt.show()

# Accidents by lighting conditions
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Light_Conditions', order=df['Light_Conditions'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Accidents by Lighting Conditions")
plt.show()
```

ðŸ”¹ Findings from Q5: This reveals how external factors impact accidents.

Q6. Vehicle and Driver Information

```python
# Most frequently involved vehicle types
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Vehicle_Type', order=df['Vehicle_Type'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Accidents by Vehicle Type")
plt.show()
```

 Findings from Q6: This will show which vehicles are more prone to accidents

Q7.Temporal Patterns

```python
# Accidents on weekdays vs weekends
df['is_weekend'] = df['Day_of_Week'].isin(['Saturday', 'Sunday'])

plt.figure(figsize=(6, 5))
sns.countplot(data=df, x='is_weekend')
plt.xticks([0, 1], ['Weekdays', 'Weekends'])
plt.title("Accidents: Weekdays vs. Weekends")
plt.show()

# Seasonal variation in accident occurrences
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Month')
plt.title("Seasonal Variation in Accidents")
plt.show()
```

 Findings from Q7: This will show peak accident times and seasonal trends.

Q8.  Contributing Factors

```python
# Top contributing factors
df['Junction_Control'].value_counts().head(10).plot(kind='barh')
plt.title("Top 10 Contributing Factors to Accidents")
plt.xlabel("Number of Accidents")
plt.ylabel("Factor")
plt.show()
```

ðŸ”¹ Findings from Q8: This will highlight key reasons for accidents.

Q9. Injury and Fatality Analysis

```python
# Distribution of injuries and fatalities among road users
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Carriageway_Hazards', order=df['Carriageway_Hazards'].value_counts().index)
plt.title("Accidents by Road User Type")
plt.show()
```

Findings from Q9: This shows which road users are most affected.

Q10. Comparative Analysis

```python
# Compare accident stats between regions
df.groupby('Urban_or_Rural_Area')['Accident_Index'].count().sort_values(ascending=False).head(10).plot(kind='barh')
plt.title("Accidents by Region")
plt.xlabel("Number of Accidents")
plt.ylabel("Region")
plt.show()
```
